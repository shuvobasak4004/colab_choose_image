```python
import cv2
import numpy as np
import tensorflow as tf
from google.colab import files

# Load the trained model
model_path = '/content/drive/MyDrive/Gradio4/thesisCnn30.h5'  # Replace with your model path        .....Change.............
model = tf.keras.models.load_model(model_path)

# Define the disease labels
disease_labels = ['Potato___Early_blight', 'Potato___healthy', 'Potato___Late_blight']   # ...............Change........

# Function to preprocess the input image
def preprocess_image(image):
    try:
        # Resize the image to 256x256 pixels (compatible with the model)
        image = cv2.resize(image, (256, 256))
        # Convert the image to RGB format
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        # Normalize the image
        image = image / 255.0
        # Expand dimensions to match model input shape
        image = np.expand_dims(image, axis=0)
        return image
    except Exception as e:
        return None

# Upload an image interactively
uploaded_files = files.upload()
if len(uploaded_files) > 0:
    # Assume the first uploaded file is the image (you can modify this if needed)
    file_name = list(uploaded_files.keys())[0]
    sample_image = cv2.imdecode(np.frombuffer(uploaded_files[file_name], np.uint8), -1)

    processed_image = preprocess_image(sample_image)

    if processed_image is not None:
        try:
            prediction = model.predict(processed_image)
            predicted_label = disease_labels[np.argmax(prediction)]
            confidence = np.max(prediction) * 100  # Multiply confidence by 100 to get percentage

            print("Predicted Disease:", predicted_label)
            print("Confidence:", confidence, "%")
        except Exception as e:
            print("Error during prediction:", str(e))
    else:
        print("Error: Unable to process the uploaded image.")
else:
    print("No image uploaded. Please upload an image to continue.")
# choose image
```
